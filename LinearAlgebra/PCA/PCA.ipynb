```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
X = digits.data
Y = digits.target
```

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
X_new = model.fit_transform(X)

print(X_new.shape)
```

```python
X_new[:5]
```

```python
plt.rcParams["figure.figsize"] = (18,10)
```

```python
fig, ax = plt.subplots()


im = ax.scatter(X_new[:,0], X_new[:,1], s=25, c=Y, cmap="plasma", edgecolor='none')
cbar = fig.colorbar(im, ax=ax,label='Digit')
```
